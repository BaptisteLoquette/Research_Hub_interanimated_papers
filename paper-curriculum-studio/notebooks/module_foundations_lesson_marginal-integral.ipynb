# Marginalization as an integral

**Objectives**

- Read p(x) as an integral over latent z
- Connect to Monte Carlo and variational views


## Marginal p(x)

The data distribution is a mixture of likelihoods under the latent prior.

$$
p(x) = \int p(x \mid z) p(z) \, dz
$$

- **p(x)**: density of observations.
- **p(x|z)**: how data is generated given latent z.
- **p(z)**: prior over latents.
- **∫ dz**: sum (integrate) over all possible z.


```
import numpy as np
# toy: sample z ~ N(0,1), x|z ~ N(z, 0.1)
z = np.random.randn(1000)
x = z + np.random.randn(1000) * 0.1
print(x.mean(), x.std())
```
